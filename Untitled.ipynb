```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(['median_house_value'],axis = 1)
y = df['median_house_value']
x_train ,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
 
```

```python
train_data = x_train.join(y_train)
test_data = x_test.join(y_test)
```

```python
train_data = train_data.join(pd.get_dummies(train_data['ocean_proximity']))
test_data = test_data.join(pd.get_dummies(test_data['ocean_proximity']))

train_data
test_data
```

```python
train_data = train_data.drop(['ocean_proximity'],axis =1)
test_data = test_data.drop(['ocean_proximity'],axis =1)
```

# train model de ML 

* regression linear

```python
from sklearn.linear_model import LinearRegression
x_train, y_train = train_data.drop(['median_house_value'], axis=1
                                  ), train_data['median_house_value']
rg = LinearRegression()
rg.fit(x_train, y_train)
```

```python
x_test, y_test = test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
rg.score(x_test,y_test)
```

* Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
frst = RandomForestRegressor()
frst.fit(x_train, y_train)

frst.score(x_test, y_test)
```

```python
# vhercher automatiquement les meilleurs hyperparametres 
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

scaler = StandardScaler()
x_train_s = scaler.fit_transform(x_train)

param_grid = {
    'n_estimators': [100, 200, 300],
    'min_samples_split': [2, 4],  # CORRIGÃ‰
    'max_depth': [None, 4, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)


grid_search.fit(x_train_s, y_train)
```

```python
grid_search.best_estimator
```

```python
y_pred = rg.predict(x_test)

# PrÃ©dire sur les donnÃ©es de test
y_pred = rg.predict(x_test)

# Afficher quelques rÃ©sultats pour comparer
for true, pred in zip(y_test[:5], y_pred[:5]):
    print(f"RÃ©el : {true:.2f} | PrÃ©dit : {pred:.2f}")


import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred, alpha=0.7)
plt.xlabel("Valeurs rÃ©elles")
plt.ylabel("Valeurs prÃ©dites")
plt.title("RÃ©el vs PrÃ©dit (RÃ©gression)")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # diagonale parfaite
plt.show()
```
